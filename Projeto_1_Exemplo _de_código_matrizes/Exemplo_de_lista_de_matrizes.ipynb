```python
%%time
#Time pega o tempo pra rodar a célula
import numpy as np

def gauss_jordan_completo(matriz_aumentada):
    """Método de Gauss-Jordan com todos os passos detalhados"""
    matriz = matriz_aumentada.astype(float)
    n = len(matriz)

    print("MATRIZ AUMENTADA INICIAL:")
    print(matriz)
    print()

    # Fase de eliminação
    for i in range(n):
        # Pivoteamento parcial
        max_row = i
        for k in range(i + 1, n):
            if abs(matriz[k, i]) > abs(matriz[max_row, i]):
                max_row = k

        if max_row != i:
            matriz[[i, max_row]] = matriz[[max_row, i]]
            print(f"Troca linha {i+1} com linha {max_row+1}:")
            print(matriz)
            print()

        # Normalizar o pivô
        pivot = matriz[i, i]
        if pivot == 0:
            raise ValueError("Sistema singular")

        matriz[i] = matriz[i] / pivot
        print(f"Linha {i+1} ÷ {pivot}:")
        print(matriz)
        print()

        # Eliminar outras linhas
        for k in range(n):
            if k != i:
                factor = matriz[k, i]
                matriz[k] = matriz[k] - factor * matriz[i]
                print(f"Linha {k+1} - {factor} × Linha {i+1}:")
                print(matriz)
                print()

    return matriz[:, -1]

# Matriz aumentada do sistema
matriz_aumentada = np.array([
    [1, 1, 1, 1, 0],
    [1, 1, 1, -1, 4],
    [1, 1, -1, 1, -4],
    [1, -1, 1, 1, 2]
])

print("="*60)
print("MÉTODO DE GAUSS-JORDAN")
print("="*60)

solucao_gj = gauss_jordan_completo(matriz_aumentada)
print("SOLUÇÃO PELO MÉTODO DE GAUSS-JORDAN:")
for i, x in enumerate(solucao_gj, 1):
    print(f"x{i} = {x:.2f}")
```

```python

#Time pega o tempo pra rodar a célula
def eliminacao_gaussiana(matriz_aumentada):
    """Método de Gauss com substituição retroativa"""
    matriz = matriz_aumentada.astype(float)
    n = len(matriz)

    print("\n" + "="*60)
    print("MÉTODO DE GAUSS (ELIMINAÇÃO GAUSSIANA)")
    print("="*60)

    print("Matriz aumentada inicial:")
    print(matriz)
    print()

    # Fase de eliminação para frente
    for i in range(n):
        # Pivoteamento parcial
        max_row = i
        for k in range(i + 1, n):
            if abs(matriz[k, i]) > abs(matriz[max_row, i]):
                max_row = k

        if max_row != i:
            matriz[[i, max_row]] = matriz[[max_row, i]]
            print(f"Troca linha {i+1} com linha {max_row+1}")

        # Eliminar abaixo do pivô
        for k in range(i + 1, n):
            factor = matriz[k, i] / matriz[i, i]
            matriz[k] = matriz[k] - factor * matriz[i]
            print(f"Linha {k+1} - {factor:.2f} × Linha {i+1}")

        print(f"Após eliminação da coluna {i+1}:")
        print(np.round(matriz, 2))
        print()

    # Fase de substituição retroativa
    x = np.zeros(n)
    for i in range(n-1, -1, -1):
        x[i] = matriz[i, -1]
        for j in range(i+1, n):
            x[i] -= matriz[i, j] * x[j]
        x[i] /= matriz[i, i]

    return x

solucao_gauss = eliminacao_gaussiana(matriz_aumentada)
print("SOLUÇÃO PELO MÉTODO DE GAUSS:")
for i, x in enumerate(solucao_gauss, 1):
    print(f"x{i} = {x:.2f}")
```

```python

def metodo_cramer(A, b):
    """Método de Cramer para resolver sistemas lineares"""
    print("\n" + "="*60)
    print("MÉTODO DE CRAMER")
    print("="*60)

    det_A = np.linalg.det(A)
    print(f"det(A) = {det_A}")

    if abs(det_A) < 1e-10:
        raise ValueError("Sistema não tem solução única")

    solucao = []
    n = len(A)

    for i in range(n):
        # Criar matriz A_i substituindo coluna i por b
        A_i = A.copy()
        A_i[:, i] = b

        det_A_i = np.linalg.det(A_i)
        x_i = det_A_i / det_A
        solucao.append(x_i)

        print(f"\nMatriz A_{i+1}:")
        print(A_i)
        print(f"det(A_{i+1}) = {det_A_i}")
        print(f"x{i+1} = {det_A_i}/{det_A} = {x_i:.2f}")

    return solucao

# Separar matriz de coeficientes e vetor de termos independentes
A = matriz_aumentada[:, :-1]
b = matriz_aumentada[:, -1]

print("Matriz de coeficientes A:")
print(A)
print("\nVetor b:")
print(b)

solucao_cramer = metodo_cramer(A, b)
print("\nSOLUÇÃO PELO MÉTODO DE CRAMER:")
for i, x in enumerate(solucao_cramer, 1):
    print(f"x{i} = {x:.2f}")
```
